```python
# Importing required libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Step 1: Load the dataset
# Assuming a dataset 'purchase_history.csv' with customer purchase information
# Sample columns: ['CustomerID', 'Annual_Spending', 'Frequency_of_Purchase', 'Items_Purchased']

data = pd.read_csv('purchase_history.csv')

# Step 2: Data Preprocessing (Dropping customer IDs, standardizing the numerical features)
X = data.drop('CustomerID', axis=1)  # Assuming 'CustomerID' is not required for clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Applying K-Means clustering
# We choose 'k' clusters (you can choose k based on domain knowledge or methods like the elbow method)
k = 4  # For example, 4 clusters
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Add the cluster labels to the original data
data['Cluster'] = clusters

# Step 4: Visualizing clusters (Optional, for 2D/3D data)
# Example for 2D data visualization (assuming 2 features like 'Annual_Spending' and 'Frequency_of_Purchase')
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis')
plt.xlabel('Annual Spending')
plt.ylabel('Frequency of Purchase')
plt.title('Customer Segments')
plt.show()

# Step 5: Save the data with cluster labels
data.to_csv('clustered_customers.csv', index=False)

print("Clustering complete! Clustered data saved to 'clustered_customers.csv'.")
```
